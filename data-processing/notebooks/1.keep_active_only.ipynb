```python
variants = pd.read_csv('../data/raw/variant.csv')
variant_property = pd.read_csv('../data/raw/variant_property.csv')
variant_file = pd.read_csv('../data/raw/variant_file.csv')
variant_affinity = pd.read_csv('../data/raw/variant_affinity.csv')
variant_image = pd.read_csv('../data/raw/variant_image.csv')
variant_option = pd.read_csv('../data/raw/variant_option.csv')
```

```python
variant_property = variant_property[
    variant_property['variant_id'].isin(variants['variant_id'])
]

variant_file = variant_file[
    variant_file['variant_id'].isin(variants['variant_id'])
]

variant_image = variant_image[
    variant_image['variant_id'].isin(variants['variant_id'])
]

variant_option = variant_option[variant_option['variant_id'].isin(variants['variant_id'])]

variant_affinity = variant_affinity[(variant_affinity['variant_id'].isin(variants['variant_id'])) & \
    (variant_affinity['affinity_group_variant_id'].isin(variants['variant_id']))]
```

```python
variants.to_csv('../data/active_only/variant.csv', index=False)
variant_property.to_csv('../data/active_only/variant_property.csv', index=False)
variant_file.to_csv('../data/active_only/variant_file.csv', index=False)
variant_image.to_csv('../data/active_only/variant_image.csv', index=False)
variant_option.to_csv('../data/active_only/variant_option.csv', index=False)
variant_affinity.to_csv('../data/active_only/variant_affinity.csv', index=False)
```
