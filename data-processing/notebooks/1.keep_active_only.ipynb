```python
import pandas as pd
variants = pd.read_csv('../data/raw/variant.csv')
variant_property = pd.read_csv('../data/raw/variant_property.csv')
variant_file = pd.read_csv('../data/raw/variant_file.csv')
variant_affinity = pd.read_csv('../data/raw/variant_affinity.csv')
variant_image = pd.read_csv('../data/raw/variant_image.csv')
variant_option = pd.read_csv('../data/raw/variant_option.csv')

print(variants.columns)
```

```python
variants['variant_id'].nunique()
```

```python



variants['lifecycle_status'].unique()
variants = variants[['variant_id', 'product_id', 'variant_name',
       'product_name', 'product_type', 'frontend_category', 'frontend_subcategory',
       'collection', 'color_tone', 'description',
       'currency', 'sale_price', 'variant_url',
       'review_count', 'review_rating']]
variants = variants[~variants['frontend_subcategory'].isin(['Swatches'])]
variants.shape
variant_property = variant_property[
    variant_property['variant_id'].isin(variants['variant_id'])
]

variant_file = variant_file[
    variant_file['variant_id'].isin(variants['variant_id'])
]

variant_image = variant_image[
    variant_image['variant_id'].isin(variants['variant_id'])
]

variant_option = variant_option[variant_option['variant_id'].isin(variants['variant_id'])]

variant_affinity = variant_affinity[(variant_affinity['variant_id'].isin(variants['variant_id'])) & \
    (variant_affinity['affinity_group_variant_id'].isin(variants['variant_id']))]
variant_property.columns
variant_property = variant_property[variant_property['property_type'].isin(['Material','Material details','Material helper'])]
variant_property = variant_property.groupby('variant_id')['property_value'].apply(lambda x: '; '.join(x)).reset_index(name='material')
print(variant_property.shape)
print(variant_property['variant_id'].nunique())
variant_option = variant_option[~variant_option['option_type'].isin(['Swatch'])]
option_types = variant_option['option_type'].unique()
option_types.sort()
option_types
variant_option
variant_option['variant_id'].nunique()
# variant_option
variant_option['combined_pair'] = variant_option['option_type'] + ': ' + variant_option['option_value'].astype(str)
variant_option = variant_option.groupby('variant_id')['combined_pair'].agg(
    lambda x: '; '.join(x)
).reset_index(name='other_properties')
print(variant_option.shape)
print(variant_option['variant_id'].nunique())
## join variants, vaiant_property, variant_option
variants = variants.merge(variant_property, on='variant_id', how='left'
).merge(
    variant_option, on='variant_id', how='left'
)
variants.columns
variants.to_csv('../data/active_only/variant.csv', index=False)
variant_property.to_csv('../data/active_only/variant_property.csv', index=False)
variant_file.to_csv('../data/active_only/variant_file.csv', index=False)
variant_image.to_csv('../data/active_only/variant_image.csv', index=False)
variant_option.to_csv('../data/active_only/variant_option.csv', index=False)
variant_affinity.to_csv('../data/active_only/variant_affinity.csv', index=False)
```
